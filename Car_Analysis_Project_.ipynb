```python
# For Replacing missing Values / Treating Missing Values

for i in df.columns: # iterating columns from dataframe
  if df[i].dtype=="O": # detecting the data type, if object then
    m=df[i].mode()[0] # modal value
    df[i]=df[i].fillna(m)
  else:
    me=df[i].mean()
    df[i]=df[i].fillna(me)
```
